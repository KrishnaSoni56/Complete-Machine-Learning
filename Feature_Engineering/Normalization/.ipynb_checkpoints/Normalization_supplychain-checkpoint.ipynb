```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("D://Feature_csv_Files/supply_chain.csv" , usecols= ["Product type" , "SKU" , "Price" , "Number of products sold" ,"Revenue generated", "Location" , "Transportation modes" , "Routes"]   )
df.head()
```

```python
lb = LabelEncoder()
```

```python
df['Product type'] = lb.fit_transform(df["Product type"])
df['SKU'] = lb.fit_transform(df["SKU"])
df['Routes'] = lb.fit_transform(df["Routes"])
df['Location'] = lb.fit_transform(df["Location"])
df["Transportation modes"] = lb.fit_transform(df["Transportation modes"])
```

```python
x = df.drop(columns = ["Revenue generated"] , axis = 1)
```

```python
y = df["Revenue generated"]
```

```python
x_train , x_test  , y_train , y_test = train_test_split(x,y , test_size= 0.2 , random_state= 34)
```

```python
sc = StandardScaler()
mn = MinMaxScaler()
```

```python
x_train_scaler = sc.fit_transform(x_train)
```

```python
np.round(x_train.describe() , 1)
```

```python
x_train_newscaler = pd.DataFrame(x_train_scaler , columns = x_train.columns)
```

```python
np.round(x_train_newscaler.describe() , 1)
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
np.round(x_train.describe(), 2)
```

```python
x_train_new_mn = pd.DataFrame(x_train_mn , columns = x_train.columns)
```

```python
np.round(x_train_new_mn.describe() , 2)
```
