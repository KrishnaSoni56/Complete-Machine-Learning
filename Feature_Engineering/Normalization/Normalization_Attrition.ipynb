```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("D:\\Feature_csv_Files\\Attrition.csv" , usecols= ["Age" , "Attrition" , "Department" , "EducationField" , "Gender" , "MaritalStatus" , "OverTime" ])
```

```python
df.head()
```

```python
lb = LabelEncoder()
```

```python
df['Attrition'] = lb.fit_transform(df['Attrition'])
df['Department'] = lb.fit_transform(df['Department'])
df['EducationField'] = lb.fit_transform(df['EducationField'])
df['MaritalStatus'] = lb.fit_transform(df['MaritalStatus'])
df['OverTime'] = lb.fit_transform(df['OverTime'])
df['Gender'] = lb.fit_transform(df['Gender'])


```

```python
x= df.drop(columns= ["Attrition"])
y = df["Attrition"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size= 0.3 , random_state=34)
```

```python
mn =MinMaxScaler()
```

```python
x_train_scaledmn = mn.fit_transform(x_train)
```
